<h2>Importing the libraries</h2>

```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import sklearn.datasets as datasets
from sklearn.model_selection import train_test_split
```

<h2>Importing the data</h2>

```python
url = "https://drive.google.com/file/d/11Iq7YvbWZbt8VXjfm06brx66b10YiwK-/view"
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
```

<h2>Dropping the ID column</h2>

```python
df.drop(index=1, columns='Id').head()
```

```python
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
X[0:5]
```

```python
y = df[['Species']]
y[0:5]
```

<h2>Setting up the decision tree</h2>

<h3><i>Creating train and test data</i></h3>

```python
from sklearn.model_selection import train_test_split
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
```

<h2>Modeling</h2>

```python
irisTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
irisTree # it shows the default parameters
```

```python
irisTree.fit(X_trainset, y_trainset)
```

<h2>Prediction</h2>

```python
predTree = irisTree.predict(X_testset)
```

```python
print (predTree [0:5])
print (y_testset [0:5])
```

<h2>Evaluation</h2>

```python
from sklearn import metrics
import matplotlib.pyplot as plt
```

<h3>For training set</h3>

```python
train_pred = irisTree.predict(X_trainset)
train_pred
```

```python
print("Accuracy score:", metrics.accuracy_score(train_pred, y_trainset))
```

<h3>For test set</h3>

```python
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predTree))
```
